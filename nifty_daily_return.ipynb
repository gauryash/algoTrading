```python

"""

@author: Yash_Gaur
"""
import pandas as pd
import matplotlib.pyplot as plt
data_link=".../nifty_data.csv"
```

```python
df=pd.read_csv(data_link)  # read csv from link and make data frame
print(df.head())
```

```python
#df=df.drop(columns=["Unnamed: 7"])         #unwanted column drop
df_close=pd.DataFrame(df,columns=["Close"]) # dataframe slice 
df['daily_return']=df_close.pct_change()          # calculate daily return
daily_return=daily_return.dropna()          # drop NA values from dataset
df['daily_cum_return']= (daily_return +1 ).cumprod()  # calculate compounded returns of underlying
df.head()
```

```python
df.set_index("Date") # set index as date
```

```python
plt.subplot(2,2,1)
plt.plot( daily_return)
plt.subplot(2,2,2)
plt.plot(daily_cum_return)
```

```python
# this figure show daily return of underlying
fig , ax = plt.subplots(figsize=(15,10))
ax.plot(df["Date"],df["daily_return"])
plt.show()
```

```python
# this figure show daily return of underlying
fig , ax = plt.subplots(figsize=(15,10))
ax.plot(df["Date"],df["daily_cum_return"])
plt.show()
```
